```python
#import the matplotlib
import matplotlib.pyplot as plt
```

```python
#Basic Line Plot


# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

# Create a simple line plot
plt.plot(x, y, marker="o", linestyle="-", color="b", label="Sales Data")

# Add labels and title
plt.xlabel("Time (Months)")
plt.ylabel("Sales (Units)")
plt.title("Sales Over Time")
plt.legend()

# Show the plot
plt.show()
```

```python
#Bar Chart


# Data
categories = ["Apple", "Banana", "Cherry", "Date"]
values = [50, 80, 30, 90]

# Create bar chart
plt.bar(categories, values, color=["red", "yellow", "pink", "brown"])

# Add labels and title
plt.xlabel("Fruits")
plt.ylabel("Quantity Sold")
plt.title("Fruit Sales in a Store")

# Show the plot
plt.show()
```

```python
#Scatter Plot (Best for Data Relationships)
import matplotlib.pyplot as plt

# Data
x = [5, 10, 15, 20, 25]
y = [50, 55, 60, 62, 70]

# Create scatter plot
plt.scatter(x, y, color="purple", marker="x", s=100)  # 's' controls marker size

# Add labels and title
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Study Time vs Exam Score")

# Show the plot
plt.show()
```

```python
#Histogram (Best for Data Distribution)
import matplotlib.pyplot as plt
import numpy as np

# Generate random data (ages)
ages = np.random.randint(18, 60, 100)

# Create histogram
plt.hist(ages, bins=8, color="green", edgecolor="black")

# Add labels and title
plt.xlabel("Age Groups")
plt.ylabel("Frequency")
plt.title("Age Distribution of Customers")

# Show the plot
plt.show()
```

```python
# Pie Chart (Best for Percentages)
import matplotlib.pyplot as plt

# Data
labels = ["Python", "Java", "C++", "JavaScript"]
sizes = [40, 25, 20, 15]
colors = ["blue", "red", "green", "orange"]

# Create pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%", startangle=140)

# Add title
plt.title("Programming Language Popularity")

# Show the plot
plt.show()
```

```python
# Subplots (Multiple Charts in One Figure)
import matplotlib.pyplot as plt

# Create figure and subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# First subplot (Line Plot)
ax[0].plot([1, 2, 3, 4], [10, 20, 25, 30], marker="o", color="blue")
ax[0].set_title("Line Chart")
ax[0].set_xlabel("X-axis")
ax[0].set_ylabel("Y-axis")

# Second subplot (Bar Chart)
ax[1].bar(["A", "B", "C", "D"], [5, 10, 15, 20], color="red")
ax[1].set_title("Bar Chart")

# Show both plots
plt.show()
```
