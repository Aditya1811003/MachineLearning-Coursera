```python
import pandas as pd

# Load dataset
df = pd.read_csv("sample_feature_scaling.csv")

# Display first few rows
print(df.head())

```

```python
# Check data types and statistics
print(df.describe())
```

```python
from sklearn.preprocessing import StandardScaler

# Select numerical columns
num_features = ["Age", "Income", "Spending_Score", "Transaction_Frequency"]

# Apply StandardScaler
scaler = StandardScaler()
df_standardized = df.copy()
df_standardized[num_features] = scaler.fit_transform(df[num_features])

# Display transformed dataset
print(df_standardized.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

# Apply MinMaxScaler
scaler = MinMaxScaler()
df_normalized = df.copy()
df_normalized[num_features] = scaler.fit_transform(df[num_features])

# Display transformed dataset
print(df_normalized.head())
```

```python
# Save Standardized Dataset
df_standardized.to_csv("standardized_feature_scaling.csv", index=False)

# Save Normalized Dataset
df_normalized.to_csv("normalized_feature_scaling.csv", index=False)

print("✅ Feature Scaling complete! Processed datasets saved.")
```
